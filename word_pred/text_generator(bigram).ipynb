```python
nlp = spacy.load('en_core_web_lg')

def preprocess (text):
    text = text.lower()
    doc = nlp(text)

    new_text = []

    for word in doc:
        if not(word.is_punct or word.like_url or word.like_email or word.is_space):
            new_text.append(word.lemma_)
    
    new_text = " ".join(new_text)
    return new_text
```

```python
processed = df[0].apply(preprocess)
```

```python
from nltk import word_tokenize

bi_gram = []

def bi_gram_collection(text):
    text = word_tokenize(text)

    for i in range(0, len(text) - 1):
        bi_gram.append(text[i:i+2])
    
    # for i in range(len(0, len(doc - 1))):
    #     bi_gram.append(doc[i])
    #     i += 1
    return bi_gram
            
```

```python
bi_gram_collection(processed[0])
```

```python
bigrams = processed.apply(bi_gram_collection)
```

```python
bigrams[0][2]
```

```python
bgrams = {}

for word in bigrams[0]:
    if(word[0] not in bgrams):
        bgrams[word[0]] = {}
    
    if(word[1] not in bgrams[word[0]]):
        bgrams[word[0]][word[1]] = 1
    else:
        bgrams[word[0]][word[1]] += 1
```

```python
bgrams
```

```python
for word in bgrams:
    bgrams[word] = (sorted(bgrams[word].items(), key=lambda item: item[1], reverse=True))
```
